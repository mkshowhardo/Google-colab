```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
```

```python
df.head(5)
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]

cols
```

```python
df[cols].head(5)
```

```python
df['education_level'].value_counts()
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df = df[cols].dropna()
```

```python
new_df.shape,df[cols].shape
```

```python
new_df.hist(bins = 50, density =True, figsize= (20,20))

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].hist(bins = 50, density = True, ax = ax, color = 'red')

new_df['training_hours'].hist(bins = 50, density = True, ax = ax, color = 'blue',alpha = 0.8)

plt.show()


```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


df['training_hours'].plot.density(color = 'red')

new_df['training_hours'].plot.density(color = 'blue')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


df['city_development_index'].plot.density(color = 'red')

new_df['city_development_index'].plot.density(color = 'blue')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


df['experience'].hist(bins = 50, density = True, ax = ax, color = 'red')

new_df['experience'].hist(bins = 50, density = True, ax = ax , color = 'blue', alpha = 0.8)
```
